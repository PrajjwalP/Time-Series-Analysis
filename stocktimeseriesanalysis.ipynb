```r
library(plotly)
```

Used the Microsoft's stock time series dataset which contains of 6 variables with a data period of 7 years. 

```r
dataset = read.csv('../input/microsoft/ms_stock_dataset.csv')
```

```r
head(dataset)
tail(dataset)
```

```r
summary(dataset)
is.null(dataset) 
```

```r
fig <- plot_ly(dataset, x = ~Date, y = ~Open, type = 'scatter', mode = 'lines')

fig
```

The above plot compares the opening price with the dates.

```r
fig <- plot_ly(dataset, type = 'scatter', mode = 'lines')%>%
  add_trace(x = ~Date, y = ~High, name = 'Volume Traded')%>%
  layout(showlegend = F)
options(warn = -1)

fig <- fig %>%
  layout(
         xaxis = list(zerolinecolor = '#ffff',
                      zerolinewidth = 2,
                      gridcolor = 'ffff'),
         yaxis = list(zerolinecolor = '#ffff',
                      zerolinewidth = 2,
                      gridcolor = 'ffff'),
         plot_bgcolor='#e5ecf6', width = 900, 
         title=list(text='High Stock Price as per Dates'))


fig
```

```r
fig <- plot_ly(dataset, type = 'scatter', mode = 'lines')%>%
  add_trace(x = ~Date, y = ~Volume, name = 'Volume Traded')%>%
  layout(showlegend = F)
options(warn = -1)

fig <- fig %>%
  layout(
         xaxis = list(zerolinecolor = '#ffff',
                      zerolinewidth = 2,
                      gridcolor = 'ffff'),
         yaxis = list(zerolinecolor = '#ffff',
                      zerolinewidth = 2,
                      gridcolor = 'ffff'),
         plot_bgcolor='#e5ecf6', width = 900,
      title=list(text='Volume Traded'))


fig
```
